## Regression of Alipur, Delhi - DPCC AQI & GVM Corporation, Vizag - APPCB AQI
## Plotting a line of best fit of Alipur, Delhi - DPCC AQI & GVM Corporation, Vizag - APPCB AQI

Plotting a line of best fit plots, the line derived from performing Ordinary Least Squares (OLS) regression on a dataset.

**m = slope, b = intercept**

```python
df = pd.read_excel('Data_Analytics.xlsx')
x = df["Alipur, Delhi - DPCC AQI"]
y = df["GVM Corporation, Vizag - APPCB AQI"]
m, b = np.polyfit(x, y, 1)
plt.style.use('seaborn')
plt.plot(x, y, 'o')
plt.plot(x, m*x + b)
```
